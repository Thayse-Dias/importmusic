```python
import pygame
import time

# Inicializa o mixer do pygame
pygame.mixer.init()

# Lista de músicas (substitua pelos caminhos das suas músicas)
playlist = [
    r"C:\Users\thays\OneDrive\Área de Trabalho\Music\Give Me Your Forever - Freen Becky The Loyal Pin Gala Premiere_[_YouConvert.net_].mp3",
    r"C:\Users\thays\OneDrive\Área de Trabalho\Music\Mood (Lofi)_[_YouConvert.net_].mp3"
]

def tocar_playlist(playlist):
    for musica in playlist:
        print(f"Tocando: {musica}")
        pygame.mixer.music.load(musica)
        pygame.mixer.music.play()
        
        # Aguarda o término da música
        while pygame.mixer.music.get_busy(): 
            time.sleep(1)  # Espera 1 segundo

if __name__ == "__main__":
    tocar_playlist(playlist)
```
